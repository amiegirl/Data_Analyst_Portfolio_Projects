## 1.) Import Libraries

```python
import requests
from scrapy import Selector
```

# Inspecting url

```python
response = requests.get('https://cheflolaskitchen.com/nigerian-recipes-you-need-to-try/')
response
```

```python
html = requests.get('https://cheflolaskitchen.com/nigerian-recipes-you-need-to-try/').content
sel = Selector(text=html)
sel
```

```python
dish = sel.xpath('/html/body/div[2]/div/main/article/div/h2/a/text()').extract()
dish
```

```python
len(dish)
```

```python
description = sel.xpath('//*[@id="post-26234"]/div/p//text()[1]').extract()[3:-5]
description
```

```python
len(description)
```

```python
description.remove('Egusi Ijebu Recipe')
description.remove('Lumpy Egusi Stew')
description.remove('Okro Soup with spinach')
```

```python
import pandas as pd
data = pd.DataFrame({'dish':dish,'description':description})
data.to_csv('meal.csv', index=False)
```
